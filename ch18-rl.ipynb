```python
import copy

def discount_reward(orig_rewards, r):
    
    all_rewards = copy.deepcopy(orig_rewards)
    
    for rewards in all_rewards:
        for ind in range(len(rewards) - 2, -1, -1):
            rewards[ind] += rewards[ind+1]*r
    
    return all_rewards

def calculate_action_advantage(all_rewards, r):
    
    action_return = discount_reward(all_rewards, r)
    
    all_actions_return = np.concatenate([act for act in action_return])
    all_actions_mean = all_actions_return.mean()
    all_actions_std = all_actions_return.std()
    
    return [list((np.array(act) - all_actions_mean)/all_actions_std) for act in action_return]
```

```python
def play_one_step(env, model, obs, policy_grad = False):
    
    ## hard code policy
    
    with tf.GradientTape() as tape:
        proba = model(obs[np.newaxis])
        action = tf.cast(tf.random.uniform([1,1]) > proba, dtype=tf.int32)
        obs, reward, done, _ = env.step(action.numpy()[0][0])
        
        if not policy_grad:
            y_target = np.array([0] if obs[2] > 0 else [1])
        else:
            y_target = tf.cast(1 - action, dtype=tf.int32)
            
        loss = tf.reduce_mean(tf.metrics.binary_crossentropy(y_target, proba))
    
    grads = tape.gradient(loss, model.trainable_variables)
    
    return obs, reward, done, grads
```

```python
obs, reward, done, grads = play_one_step(env, model, obs, True)
```

```python
optimizer = keras.optimizers.Adam()

def play_episode(env, model, update_on_eps_end = True, max_iter = 100):
    
    done = False
    rewards = []
    obses = []
    frames = []
    grads = []
    
    obs = env.reset()
    iter_ = 0
    
    while not done and iter_ < max_iter:
        
        ## render img
        frames.append(env.render(mode='rgb_array'))
        
        obs, reward, done, grad = play_one_step(env, model, obs)
        
        if update_on_eps_end:
            optimizer.apply_gradients(zip(grads, model.trainable_variables))
        
        rewards.append(reward)
        obses.append(obs)
        grads.append(grad)
        
        iter_ += 1
    
    ## render final img
    frames.append(env.render(mode='rgb_array'))
        
    return rewards, obses, frames, grads
```

```python
rewards, obses, frames, grads = play_episode(env, model)
```

```python
def train_model(env, model, n_iter):
    
    all_frames = []
    all_obses = []
    all_rewards = []
    
    for i in range(n_iter):
        rewards, obses, frames, grads = play_episode(env, model)
        
        all_frames.append(frames)
        
    return all_rewards, all_obses, all_frames
```

## Now, let's implement policy gradient

Policy gradient have 4 steps as follows:  
1. Play many episodes to get many of rewards and gradients for each episode. (Not yet update gradients) 
2. Calculate action's return using predefine discount factor  
3. Calculate action's advantage  
4. update gradient based on action's advantage value (element wise multiplication between action's advantage and gradient, then update the parameter using average gradient)

```python
def cal_mean_grads(all_rewards, all_grads, discount_factor):
    
    action_advantage = calculate_action_advantage(all_rewards, discount_factor)
    flatten_grads = [[] for i in range(len(all_grads[0][0]))]
    
    for episode, grads in enumerate(all_grads):
        # grads of each episode
        for action, grad in enumerate(grads):
            # grad of each action
            for ind in range(len(grad)):
                grad[ind] = grad[ind]*action_advantage[episode][action]
                flatten_grads[ind].append(grad[ind])
    
#     print('flatten grads')
#     print(tf.shape(tf.stack(flatten_grads[1])))
#     print('test reduce mean')
#     print(tf.reduce_mean(tf.stack(flatten_grads[1]), axis = 0))
    
    return tf.reduce_mean(tf.stack(flatten_grads[1]), axis = 0)
```

```python
keras.backend.clear_session()
```

```python
def train_policy_grad(env, model, n_iter, n_episode, discount_factor = 0.9, optimizer = keras.optimizers.RMSprop()):

    for i in range(n_iter):
        
        all_frames = []
        all_obses = []
        all_rewards = []
        all_grads = []
        
        for num_eps in range(n_episode):
            rewards, obses, frames, grads = play_episode(env, model, update_on_eps_end = False, max_iter = 300)
            
            all_grads.append(grads)
            all_rewards.append(rewards)
            all_frames.append(frames)
            
        mean_grads = cal_mean_grads(all_rewards, all_grads, discount_factor)
        optimizer.apply_gradients(zip(mean_grads, model.trainable_variables))
        
    return all_frames, all_obses, all_rewards, all_grads
```
